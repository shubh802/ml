### Missing Values

```python
# Check the percentage of nan values in each features
# The list of features which has missing values

features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() > 0]

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4),'% missing values' )
```

#### Find the relationship between the missing values and Sales Price

```python
for feature in features_with_na:
    data = dataset.copy()
    # The null values is converted to 1 and rest all as 0
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

We need to replace the NaN values with something meaningful in Feature Engineering as there is the relation between the NaN and dependent varaibles

### Numerical Variables

```python
#list of numerical variables, O means obj str field values
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print("Numerical variables: ",len(numerical_features))
dataset[numerical_features].head()
```

#### Temporal Variables(Datetime Variable)
From the dataset we have 4 year variables. We have to extract information from the year varaibles

```python
# Variables that contain year information 
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# Explore the content of these variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# We are trying to find out the relation between the YrSold and the SalesPrice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median of Sales Price')
plt.title("House Price vs year Sold")
```

```python
year_feature
```

```python
# Compare difference with all year feature with SalePrice
for feature in year_feature:
    if feature!= 'YrSold':
        data=dataset.copy()
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
dataset[numerical_features].head()
```

```python
# Numerical values are of 2 types
# 1.Continuos Variable 2.Discrete Varaible
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete variable count:{}".format(len(discrete_feature)))
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```

#### Continous Variables

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continous feature: {}".format(len(continous_feature)))
continous_feature
```

```python
for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# Our data is skewed so we will use the log transformation for Gaussian Distribution

for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

### Outliers


```python
for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
