### COVID Tracking - India

```python
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

#### Total positive case in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align='center', alpha=0.5,
            color=(1,0,0),
            edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+100)
plt.xlabel('Total positive cases')
plt.title('Corona Virus Cases')
```

#### Total Active Cases In India

```python
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['death'], align='center', alpha=0.5,
            color=(1,0.5,0),
            edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+100)
plt.xlabel('Active No of Cases')
plt.title('Corona Virus Cases')
```

#### Total no of Deaths

```python
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['death'], align='center', alpha=0.5,
            color=(0.4,0.5,0),
            edgecolor=(0.4,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+100)
plt.xlabel('Total no of Deaths')
plt.title('Corona Virus Cases')
```
